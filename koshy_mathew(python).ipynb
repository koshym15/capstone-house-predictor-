#### let's check for multicollinearity with VIF

```python
#let's first concat numerical and categorical variables 
df2 = pd.concat([df_num2 ,df_cat2, df_cat1['zipcode']] , axis =1)
```

```python
df2.info()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = df2.drop(['price' ], axis = 1)
calc_vif(X).sort_values(by = 'VIF', ascending = False , ignore_index = True)
```

```python
X = df2.drop(['price', 'lot_measure' ], axis = 1)
calc_vif(X).sort_values(by = 'VIF', ascending = False)
```

```python
from sklearn.preprocessing  import StandardScaler
```

```python
std = StandardScaler()
df3 = pd.DataFrame(std.fit_transform(df2) , columns= df2.columns)
```

```python
df3.head()
```

```python
# using kmeans clustering , let's first find optimal cluster using ws plot
from sklearn.cluster import KMeans
ws =[]
for i in range(1,11):
    KM = KMeans(n_clusters=i)
    KM.fit(df3)
    ws.append(KM.inertia_)
```

```python
plt.plot(range(1,11), ws)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df3)
```

```python
df2['clusters'] = kmeans.predict(df3)
```
